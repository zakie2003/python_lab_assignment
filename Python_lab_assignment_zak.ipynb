# **Q.4)Program to check wheter the number is prime or not**

```python
num=int(input("Enter the number:"))
if num > 1:
    # Iterate from 2 to n / 2
    for i in range(2, int(num/2)+1):
        # If num is divisible by any number between
        # 2 and n / 2, it is not prime
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

# **Q.25)Write a Program to add 2 matrice**

# **Q.26) Write a Program to multiply 2 matrice**

```python
m1=[]
m2=[]
m3=[]

col=int(input("Enter columns: "));
row=int(input("Enter rows: "));

for a in range(row):
    Col=[]
    for b in range(col):
        Col.append(int(input("Enter the element: ")));
    m1.append(Col);

for a in range(row):
    Col=[]
    for b in range(col):
        Col.append(int(input("Enter the element: ")));
    m2.append(Col);

for a in range(row):
    Col=[]
    for b in range(col):
        Col.append(0);
    m3.append(Col);

for a in range(row):
    for b in range(col):
        for c in range(row):
            m3[a][b]+=m1[a][c]*m2[c][b];

for a in range(row):
    for b in range(col):
        print(m3[a][b],end=" ");
    print();
```

# **Q.38) Program to remove punctuations from a string**

# **Q.41)Program to check if a string is anagram or not.**

```python
s=input("Enter the string 1:")

r=input("Enter the String 2:")
l1=[]
l2=[]
for i in s:
  l1.append(i)

for i in r:
  l2.append(i)

l1.sort()
l2.sort()
if("".join(l1)=="".join(l2)):
  print(f"{s} is an anagram of {r}")
```

# **Q.43)Program to reverse all the words in a given sentance**

```python
class Sen:
  def __init__(self,s=""):
    self.s=s

  def r(self):
    l=self.s.split(" ")
    l=l[::-1]
    return " ".join(l)
n=input("Enter the sentance:")
s1=Sen(n)
result=s1.r()
print("The reverse of the given sentance:")
print(result)
```
